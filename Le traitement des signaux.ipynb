```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Paramètres du signal
frequence = 5  # en Hertz
amplitude = 1  # amplitude du signal
phase = 0  # déphasage
temps = np.linspace(0, 1, 1000)  # intervalle de temps de 0 à 1 seconde

# Générer le signal sinusoïdal
signal = amplitude * np.sin(2 * np.pi * frequence * temps + phase)
```

```python
# Tracer le signal
plt.figure(figsize=(10, 4))
plt.plot(temps, signal)
plt.title('Signal sinusoïdal')
plt.xlabel('Temps (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
frequence_echantillonnage = 500  # fréquence d'échantillonnage en Hz
# Calcul de la FFT
fft = np.fft.fft(signal)
frequences = np.fft.fftfreq(len(signal), 1 / frequence_echantillonnage)

# Prendre la moitié du spectre car la FFT est symétrique
fft = fft[:len(fft) // 2]
frequences = frequences[:len(frequences) // 2]
# Tracer le spectre de fréquence
plt.subplot(2, 1, 2)
plt.plot(frequences, np.abs(fft))
plt.title('Transformée de Fourier (FFT)')
plt.xlabel('Fréquence (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Ajouter du bruit gaussien (bruit blanc)
intensite_bruit = 0.1  # Intensité du bruit
bruit = intensite_bruit * np.random.normal(size=temps.shape)

# Signal avec bruit
signal_bruite = signal + bruit
```

```python
plt.figure(figsize=(10, 4))
plt.plot(temps, signal_bruite, label='Signal avec bruit')
#plt.plot(temps, signal, label='Signal original', linestyle='--')
plt.title('Signal sinusoïdal avec bruit')
plt.xlabel('Temps (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
frequence_echantillonnage = 500  # fréquence d'échantillonnage en Hz
# Calcul de la FFT
fft = np.fft.fft(signal_bruite)
frequences = np.fft.fftfreq(len(signal_bruite), 1 / frequence_echantillonnage)

# Prendre la moitié du spectre car la FFT est symétrique
fft = fft[:len(fft) // 2]
frequences = frequences[:len(frequences) // 2]
# Tracer le spectre de fréquence
plt.subplot(2, 1, 2)
plt.plot(frequences, np.abs(fft))
plt.title('Transformée de Fourier (FFT)')
plt.xlabel('Fréquence (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
